```python
import pandas as pd
import os
import datetime
```

```python
class DataCleaner():
    
    def __init__(self):
        pass
    
    def convert_df(self, df, inplace=False):
        # replace whitespaces in columns with underscores
        df.columns = [col.replace(' ', '_') for col in df.columns]
        
        # convert datestring to dates 
        df['Date'] = df['Date'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
        df['Year'] = df['Date'].apply(lambda x: x.year)
        df['Month'] = df['Date'].apply(lambda x: x.month)
        df['Day'] = df['Date'].apply(lambda x: x.day)
        df.drop('Date', axis=1, inplace=True)
        return df
```

```python
cleaner = DataCleaner()
```

```python
data_dir = '../../Data'
for file in os.listdir(data_dir):
    if file.endswith(".csv"):
        csv_file = os.path.join(data_dir, file)
        
        # train a model and save it in directory "models"
        # will later on be used in an ensemble predictor
        print(csv_file)
```

```python
df = pd.read_csv('../../Data/MMM.csv')
df = cleaner.convert_df(df)
```
