### Stock History

```python
google_df[['Open', 'Close', 'High', 'Low']].plot(figsize=(12,12))
plt.title('Stock History of Google')
plt.ylabel('Price per Share in USD')
plt.xlabel('Time')
plt.show()
```

doubled price since 2014!

### Volume traded

```python
google_df['Volume'].hist()
plt.show()
```

- mostly 200k volume per day

### Correlations

```python
import seaborn as sns
corr = google_df[['Open', 'High', 'Low', 'Close', 'Year', 'Month', 'Day']].corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

- of course, prices are highly correlated
- the year is higher correlated than month or day/ this makes sense, as the year has a higher overall aggregation (long time-effect)
